```python
import duckdb
import os

# Connect to the database file
con = duckdb.connect('../duckdb/olist.duckdb')

# Get a list of all CSV files you just moved
csv_files = [f for f in os.listdir('data') if f.endswith('.csv')]

for file in csv_files:
    # Create a table name by removing '.csv' (e.g., 'olist_orders_dataset')
    table_name = file.replace('.csv', '')
    
    # Ingest each file into its own table
    con.execute(f"""
        CREATE TABLE IF NOT EXISTS {table_name} AS 
        SELECT * FROM read_csv_auto('data/{file}');
    """)
    print(f"Successfully ingested: {table_name}")

# Now, verify by showing all tables
display(con.execute("SHOW TABLES").df())
```
